```python
import requests
import json
from xml.etree import ElementTree as Et
import xmltodict
```

```python
tally_url = "http://localhost:9000"
headers = {"Content-Type": "application/xml"}

xml_request = """
<ENVELOPE>
	<HEADER>
		<VERSION>1</VERSION>
		<TALLYREQUEST>EXPORT</TALLYREQUEST>
		<TYPE>COLLECTION</TYPE>
		<ID>StockItems</ID>
	</HEADER>
	<BODY>
		<DESC>
			<STATICVARIABLES>
				<SVEXPORTFORMAT>$$SysName:XML</SVEXPORTFORMAT>
			</STATICVARIABLES>
			<TDL>
				<TDLMESSAGE>
					<COLLECTION Name='StockItems'>
						<TYPE>StockItem</TYPE>
						<FETCH>Name,GUID,Parent,ClosingBalance,BaseUnits,LastPurcParty,OpeningBalance</FETCH>
					</COLLECTION>
				</TDLMESSAGE>
			</TDL>
		</DESC>
	</BODY>
</ENVELOPE>
"""

response = requests.post(tally_url, data=xml_request, headers=headers)

if response.status_code == 200:
    print(response.text)
else:
    print("Failed to get response from Tally. Status code:", response.status_code)
```

```python
data_dict = xmltodict.parse(response.text)
print(json.dumps(data_dict, indent=4))  
with open('data.json', 'w') as f:
    json.dump(data_dict, f, indent=4)
```
